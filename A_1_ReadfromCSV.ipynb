```python
# (6)Have an array which contains values like "OpenCV", "Python", and 
# write a function that would determine whether any of the words from the array is present in the content, 
# if it is present then that word will be stored in another column.

# Sample array with words: OpenCV, Python, submit, deadline, tomorrow
array = ["OpenCV", "Python", "submit", "deadline", "tomorrow"]

def ContainsWord(array=array, df=dataframe, column="Raw_Message"):
    
    df['Contains'] = "-"
    
    for i in array:
        df.loc[df.loc[:, column].str.contains(i, case=False),'Contains'] = i
    
    return(df)
```

A sample array with words (OpenCV, Python, submit, deadline, tomorrow) has been defined.
the function ContainsWord() determines whether the words from the array is present in the content (Raw_Message column).
